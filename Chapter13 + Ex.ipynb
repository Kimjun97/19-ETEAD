```python
#13.3 Word histogram
import string
def process_file(filename): # 단어를 읽는 함수
    hist = dict()
    fp = open(filename) # 이 파일을 오픈한다.
    for line in fp: 
        process_line(line, hist)
    return hist
def process_line(line, hist):
    line = line.replace('-', ' ') # '-'을 ' '로 대체하게 하고 이것을 다시 라인에 넣는다. 

    for word in line.split():
        word = word.strip(string.punctuation + string.whitespace) 
        word = word.lower() # 안어를 소문자로 전환한다.
        hist[word] = hist.get(word, 0) + 1
hist = process_file("C:\\Users\\samsung\\Desktop\\emma.txt")

process_file("C:\\Users\\samsung\\Desktop\\emma.txt")
```

```python
def total_words(hist):
    return sum(hist.values())
def different_words(hist):
    return len(hist)
print('Total number of words:', total_words(hist)) # 파일의 총 단어 수
print('Number of different words:', different_words(hist)) # 서로 다른 단어 수 
```

```python
# 13.4 Most common words
def most_common(hist): # 빈번하게 나오는 단어들
    t = []
    for key, value in hist.items():# 히스트에 나온 항목들을 키값과 벨류를 측정
        t.append((value, key)) #이걸 t에 추가하고
    t.sort(reverse=True) #차순으로 정리를 한다.
    return t
t = most_common(hist) # 가장 흔한 단어를 t로 넣고
print('The most common words are:') #이 문자열과 함께
for freq, word in t[:10]: # 1등부터 10등가지 하여
    print(word, freq, sep='\t') # 출력
```

```python
# 13.5 Optional parameters

def print_most_common(hist, num=10): #이런식으로 더 간단하게 만들 수 있음
    t = most_common(hist)
    print('The most common words are:')
    for freq, word in t[:num]:
        print(word, freq, sep='\t')
        
print_most_common(hist, 20) # 20 순위까지 나타남
```

```python
def subtract(d1, d2):
    res = dict()
    for key in d1: # d1에 키를 불러오고
        if key not in d2: # 이게 만약 d2안에 없을 경우
            res[key] = None # 논
    return res # 그리고 res를 리턴

words = process_file("C:\\Users\\samsung\\Desktop\\words.txt")
diff = subtract(hist, words)
print("Words in the book that aren't in the word list:")
for word in diff:
    print(word, end=' ') # 없는 단어마다 나타낸다. ' '을 기준으로 
```

```python
# Ex 13.6
def subtract(d1, d2):
    return set(d1) - set(d2)


def main():
    hist = process_file("C:\\Users\\samsung\\Desktop\\emma.txt", skip_header=True)
    words = process_file("C:\\Users\\samsung\\Desktop\\words.txt", skip_header=False)

    diff = subtract(hist, words)
    print("The words in the book that aren't in the word list are:")
    for word in diff:
        print(word, end=' ')


```

```python
# 13.7 Random words
import random
import string

def random_word(h):
    t = []
    for word, freq in h.items():
        t.extend([word] * freq)
    return random.choice(t)
```
